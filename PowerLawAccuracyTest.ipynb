```python

import numpy as np

from tqdm import tqdm_notebook as tqdm

import matplotlib
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline  
```

```python
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [5,5]
matplotlib.rcParams.update({'font.size': 16})
```

```python
vgg11 = [88.746,  2.32 ,  2.17 ,  2.83 ]
vgg11_bn =[ 89.818, 2.07, 1.95 ,  2.99 ] 
vgg16 = [ 90.354, 2.3 ,  2.18 ,2.09 ]
vgg16_bn = [ 91.608, 2.05 ,1.97 , 3.03]
vgg19 = [ 90.822, 2.27 , 2.19 , 2.07]
vgg19_bn = [ 92.066, 2.04, 1.98 , 3.03 ] 
#vgg13 = [89.264]
#vgg13_bn = [90.494]
```

```python
accs = []
avgs = []
lyr_1_s = []
for m in [vgg11, vgg11_bn , vgg16, vgg16_bn ,vgg19 ,vgg19_bn ]:
    accs.append(m[0])
    avgs.append(np.average(m[1:3]))
    lyr_1_s.append(m[1])
   
```

```python
plt.scatter(accs, avgs, label="average")
#plt.scatter(accs, lyr_1_s, label="layer 1")

plt.title("VGG Models: Accuracies vs \n"+r"Average Power Law Exponents($\alpha$)")
plt.xlabel("accuracy")
plt.ylabel(r"average ($\alpha$)")
#plt.legend()
```

```python
accs11 = [vgg11[0], vgg11_bn[0]]
accs16 = [vgg16[0], vgg16_bn[0]]
accs19 = [vgg19[0], vgg19_bn[0]]

alphas11 = [np.average(vgg11[1:3]), np.average(vgg11_bn[1:3])]
alphas16 = [np.average(vgg16[1:3]), np.average(vgg16_bn[1:3])]
alphas19 = [np.average(vgg19[1:3]), np.average(vgg19_bn[1:3])]
```

```python
plt.scatter(accs11, alphas11, label="VGG 11")
plt.scatter(accs16, alphas16, label="VGG 16")
plt.scatter(accs19, alphas19, label="VGG 19")

#plt.scatter(accs, lyr_1_s, label="layer 1")

plt.title("VGG Models: Accuracies vs \n"+r"Average Power Law Exponents($\alpha$)")
plt.xlabel("accuracy")
plt.ylabel(r"average ($\alpha$)")
plt.legend()
```
